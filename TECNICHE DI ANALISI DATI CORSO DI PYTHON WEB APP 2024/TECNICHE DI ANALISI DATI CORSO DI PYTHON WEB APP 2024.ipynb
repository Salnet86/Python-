```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dati simulati per i fiumi e la concentrazione di sostanze chimiche
data = {
    'Fiume': ['Po', 'Arno', 'Tevere', 'Adige', 'Adda', 'P', 'Ticino', 'Magra', 'Biferno', 'Tevere'],
    'Nitrati (NO) [g/L]': [30, 60, 45, 25, 35, 50, 40, 55, 20, 48],
    'Fosfati (PO) [g/L]': [5, 12, 8, 4, 6, 9, 7, 11, 3, 8]
}

# Creiamo il DataFrame
df = pd.DataFrame(data)

# Riformattiamo i dati per un grafico a barre con Seaborn
df_melted = df.melt(id_vars='Fiume', value_vars=['Nitrati (NO) [g/L]', 'Fosfati (PO) [g/L]'], 
                    var_name='Sostanza Chimica', value_name='Concentrazione')

# Creiamo il grafico a barre
plt.figure(figsize=(12, 6))
sns.barplot(x='Fiume', y='Concentrazione', hue='Sostanza Chimica', data=df_melted)

# Aggiungiamo un titolo e le etichette per gli assi
plt.title('Concentrazione di Nitrati e Fosfati nei Fiumi Italiani')
plt.xlabel('Fiume')
plt.ylabel('Concentrazione (g/L)')
plt.xticks(rotation=45)  # Ruota le etichette dei fiumi per una migliore visibilit
plt.tight_layout()  # Assicura che le etichette non si sovrappongano

# Mostriamo il grafico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dati simulati per i fiumi e la concentrazione di ferro e alluminio
data = {
    'Fiume': ['Po', 'Arno', 'Tevere', 'Adige', 'Adda', 'P', 'Ticino', 'Magra', 'Biferno', 'Tevere'],
    'Ferro (Fe) [g/L]': [150, 200, 180, 120, 160, 190, 140, 210, 130, 170],
    'Alluminio (Al) [g/L]': [35, 45, 40, 25, 30, 50, 38, 55, 22, 42]
}

# Creiamo il DataFrame
df = pd.DataFrame(data)

# Riformattiamo i dati per un grafico a barre con Seaborn
df_melted = df.melt(id_vars='Fiume', value_vars=['Ferro (Fe) [g/L]', 'Alluminio (Al) [g/L]'], 
                    var_name='Materiale', value_name='Concentrazione')

# Creiamo il grafico a barre
plt.figure(figsize=(12, 6))
sns.barplot(x='Fiume', y='Concentrazione', hue='Materiale', data=df_melted)

# Aggiungiamo un titolo e le etichette per gli assi
plt.title('Concentrazione di Ferro e Alluminio nei Fiumi Italiani')
plt.xlabel('Fiume')
plt.ylabel('Concentrazione (g/L)')
plt.xticks(rotation=45)  # Ruota le etichette dei fiumi per una migliore visibilit
plt.tight_layout()  # Assicura che le etichette non si sovrappongano

# Mostriamo il grafico
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dati simulati per i vulcani e i livelli di emissione chimica (SO2, CO2, CO)
data = {
    'Vulcano': ['Vulcano A', 'Vulcano B', 'Vulcano C', 'Vulcano D', 'Vulcano E'],
    'SO (ppm)': [350, 300, 500, 450, 400],
    'CO (ppm)': [800, 900, 850, 1000, 950],
    'CO (ppm)': [50, 60, 70, 65, 55]
}

# Creiamo il DataFrame
df = pd.DataFrame(data)

# Riformattiamo i dati per un grafico a torrente (stacked area plot)
df.set_index('Vulcano', inplace=True)

# Creiamo il grafico a torrente (stacked area plot)
plt.figure(figsize=(12, 6))
df.plot(kind='area', stacked=True, cmap='viridis', alpha=0.7)

# Aggiungiamo un titolo e le etichette per gli assi
plt.title('Emissions of SO, CO, and CO from Volcanoes')
plt.xlabel('Vulcano')
plt.ylabel('Concentrazione (ppm)')
plt.xticks(rotation=45)  # Ruota le etichette dei vulcani per una migliore visibilit
plt.tight_layout()  # Assicura che le etichette non si sovrappongano

# Mostriamo il grafico
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Impostiamo una semina per la riproducibilit
np.random.seed(42)

# Numero di righe (100)
num_records = 100

# Generiamo dati per prodotti chimici infiammabili: Etanolo (EtOH), Benzina, Acetone, Metanolo (MeOH)
etanolo = np.random.uniform(20, 1000, num_records)  # Valori tra 20 e 1000 g/L
benzina = np.random.uniform(50, 800, num_records)  # Valori tra 50 e 800 g/L
acetone = np.random.uniform(10, 500, num_records)  # Valori tra 10 e 500 g/L
metanolo = np.random.uniform(5, 300, num_records)  # Valori tra 5 e 300 g/L

# Creiamo un DataFrame con i dati simulati
data = {
    'Tempo': pd.date_range(start='2025-01-01', periods=num_records, freq='D'),  # Periodo giornaliero
    'Etanolo (EtOH) [g/L]': etanolo,
    'Benzina [g/L]': benzina,
    'Acetone [g/L]': acetone,
    'Metanolo (MeOH) [g/L]': metanolo
}

df = pd.DataFrame(data)

# Visualizziamo le prime righe del DataFrame
print(df.head())

# Creiamo il grafico a barre
plt.figure(figsize=(14, 7))

# Grafico a barre per Etanolo (EtOH)
plt.bar(df['Tempo'], df['Etanolo (EtOH) [g/L]'], label='Etanolo (EtOH)', color='red', width=0.8)

# Grafico a barre per Benzina
plt.bar(df['Tempo'], df['Benzina [g/L]'], label='Benzina', color='blue', width=0.8)

# Grafico a barre per Acetone
plt.bar(df['Tempo'], df['Acetone [g/L]'], label='Acetone', color='green', width=0.8)

# Grafico a barre per Metanolo (MeOH)
plt.bar(df['Tempo'], df['Metanolo (MeOH) [g/L]'], label='Metanolo (MeOH)', color='purple', width=0.8)

# Aggiungiamo un titolo e le etichette per gli assi
plt.title('Concentrazione di Prodotti Chimici Infiammabili nel Tempo')
plt.xlabel('Tempo')
plt.ylabel('Concentrazione (g/L)')

# Ruotiamo le etichette dei tempi per migliorarne la visibilit
plt.xticks(rotation=45)

# Aggiungiamo la legenda
plt.legend()

# Mostriamo il grafico
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Impostiamo una semina per la riproducibilit
np.random.seed(42)

# Dati di esempio con prodotti chimici tossici (valori casuali)
prodotti = ['Cianuro', 'Fosforo', 'Cloro', 'Acido Solforico', 'Mercurio']
concentrazioni = np.random.uniform(10, 500, 5)  # Valori tra 10 e 500 g/L per ciascun prodotto chimico

# Creiamo un DataFrame
data = {
    'Prodotti Tossici': prodotti,
    'Concentrazione (g/L)': concentrazioni
}

df = pd.DataFrame(data)

# Visualizza il DataFrame
print(df)

# Creiamo un grafico a torta per visualizzare la distribuzione
plt.figure(figsize=(7, 7))

# Grafico a torta
plt.pie(df['Concentrazione (g/L)'], labels=df['Prodotti Tossici'], autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'purple', 'orange'])

# Aggiungiamo un titolo
plt.title('Distribuzione dei Prodotti Tossici')

# Mostriamo il grafico
plt.axis('equal')  # Per rendere il grafico a torta un cerchio perfetto
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Dati JSON con i prodotti e la percentuale di infiammabilit
json_data = """
{
  "prodotti": [
    {"nome": "Etanolo", "infiammazione": 75},
    {"nome": "Benzina", "infiammazione": 95},
    {"nome": "Acetone", "infiammazione": 85},
    {"nome": "Metanolo", "infiammazione": 80},
    {"nome": "Butano", "infiammazione": 90}
  ]
}
"""

# Carica i dati JSON
data = json.loads(json_data)

# Estrai i nomi dei prodotti e le percentuali di infiammabilit
prodotti = [item['nome'] for item in data['prodotti']]
infiammazione = [item['infiammazione'] for item in data['prodotti']]

# Creiamo un grafico a torta
plt.figure(figsize=(8, 8))
plt.pie(infiammazione, labels=prodotti, autopct='%1.1f%%', startangle=140, colors=['red', 'blue', 'green', 'purple', 'orange'])

# Aggiungiamo un titolo
plt.title('Percentuale di Infiammabilit dei Prodotti Infiammabili')

# Mostriamo il grafico
plt.show()
```

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dati JSON con i prodotti e la percentuale di infiammabilit
json_data = """
{
  "prodotti": [
    {"nome": "Etanolo", "infiammazione": 75},
    {"nome": "Benzina", "infiammazione": 95},
    {"nome": "Acetone", "infiammazione": 85},
    {"nome": "Metanolo", "infiammazione": 80},
    {"nome": "Butano", "infiammazione": 90}
  ]
}
"""

# Carica i dati JSON
data = json.loads(json_data)

# Creiamo un DataFrame Pandas
df = pd.DataFrame(data['prodotti'])

# Impostiamo lo stile di Seaborn
sns.set(style="whitegrid")

# Creiamo il grafico a torta con Matplotlib (utilizzando Seaborn per lo stile)
plt.figure(figsize=(8, 8))
plt.pie(df['infiammazione'], labels=df['nome'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3", len(df)))

# Aggiungiamo un titolo
plt.title('Percentuale di Infiammabilit dei Prodotti Infiammabili (Alcol e altri)', fontsize=16)

# Mostriamo il grafico
plt.show()
```
